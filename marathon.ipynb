```python
# athletes run slower in summer or winter 

#Spring → Months 3, 4, 5
#Summer → Months 6, 7, 8
#Fall (Autumn) → Months 9, 10, 11
#Winter → Months 12, 1, 2
```

```python
df2['race_month'] = df2['race_day'].str.split('.').str.get(1).astype(int)
```

```python
def assign_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
    else:
        return 'Winter'

df2['race_season'] = df2['race_month'].apply(assign_season)
```

```python
df2.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)  
```
